```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier
import random
```

## 3.3 Trees on the Banana Dataset

### 1. modify code on iris dataset and apply them on banana dataset

```python
# Load data
train_data = np.genfromtxt('data/banana_train.csv', delimiter=',')
train_y = train_data[:,0]
train_X = train_data[:,1:]
train_y = np.array([0 if y == -1 else y for y in train_y])
```

```python
# Load test data
test_data = np.genfromtxt('data/banana_test.csv', delimiter=',')
test_y = test_data[:,0]
test_X = test_data[:,1:]
test_y = np.array([0 if y == -1 else y for y in test_y])
```

```python
# Shuffle
idx = np.arange(train_X.shape[0])
np.random.seed(29)
np.random.shuffle(idx)
train_X = train_X[idx]
train_y = train_y[idx]
```

```python
# Standardize
mean = train_X.mean(axis=0)
std = train_X.std(axis=0)
train_X = (train_X - mean) / std
```

```python
# Standardize test data
test_X = (test_X - mean) / std
```

```python
# Parameters
n_classes = 2
plot_colors = "by"
plot_step = 0.02

# Train
clf = DecisionTreeClassifier().fit(train_X, train_y)

# Plot the decision boundary
x_min = train_X[:, 0].min() - 1
x_max = train_X[:, 0].max() + 1
y_min = train_X[:, 1].min() - 1
y_max = train_X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
cs2 = plt.contour(cs, levels=cs.levels[::2], color='r')

plt.axis("tight")


# Plot the training points
for i, color in zip(range(n_classes), plot_colors):
    idx = np.where(train_y == i)
    plt.scatter(train_X[idx, 0], train_X[idx, 1], c=color,
                cmap=plt.cm.Paired)

plt.axis("tight")

plt.savefig('3_3_1.png', dpi=300)
```

### 2. try different depth from 1 through 10 of decision trees

### 3. plot train and test errors as a function of depth

```python
train_errors = []
test_errors = []
for depth in range(1,11):
    # Train
    clf = DecisionTreeClassifier(max_depth=depth).fit(train_X, train_y)
    train_accu = clf.score(train_X, train_y)
    test_accu = clf.score(test_X, test_y)
    train_errors.append(1-train_accu)
    test_errors.append(1-test_accu)
    
plt.plot(range(1,11), train_errors)
plt.plot(range(1,11), test_errors)
plt.legend(['Train Errors','Test Errors'])
plt.xlabel('Tree Depth')
plt.ylabel('Error')
plt.title("Train and Test Errors with different depth of Trees")
plt.savefig('3_3_3.png', dpi=300)
print "The smallest test error is {}".format(min(test_errors))
```

### From the above plot, I got that:  1) With the depth going deeper, at the begining, both train and test errors go down from 0.40 to around 0.15;  2) However, when depth is larger than 6, train errors keep going down but test errors keep stable and even going up somewhat when the depth increases further (this may be a sign for overfitting).

### 4. Experiment with the other hyperparameters and find the combination giving the smallest test error.

```python
train_errors = []
test_errors = []
leaf_list = [1,2,5,10, 15,20,30]
for leaf in leaf_list:
    # Train
    clf = DecisionTreeClassifier(max_depth=10, min_samples_leaf=leaf).fit(train_X, train_y)
    train_accu = clf.score(train_X, train_y)
    test_accu = clf.score(test_X, test_y)
    train_errors.append(1-train_accu)
    test_errors.append(1-test_accu)
    
plt.plot(leaf_list, train_errors)
plt.plot(leaf_list, test_errors)
plt.legend(['Train Errors','Test Errors'])
plt.xlabel('Tree Minimum Leaf')
plt.ylabel('Error')
plt.title("Train and Test Errors with different Tree Minimum Leaf")
plt.savefig('3_3_4.png', dpi=300)
print "The smallest test error is {}".format(min(test_errors))
```

### As we can see from question 3, when the depth goes deeper, the tree may perfect fit the training data but perform not very well on testing data due to the overfitting problem. In order to gaurantee the generalization of tree, I tried to increase the minimum samples leaf of a tree. And as the above plot shows, when minimum samples leaf increases, the overfitting of training dataset decreases, and the testing dataset error decreases as well. Specifically, with tree depth equals to 10, when minimum samples leaf is 10, the test error reaches the optimal, which is 0.112. 

## 5. AdaBoost

## 5.1 Implementation

```python
train_y = np.array([-1 if y == 0 else y for y in train_y])
```

```python
test_y = np.array([-1 if y == 0 else y for y in test_y])
```

```python
train_y
```

```python
class AdaBoost():
    def __init__(self, train_X, train_y):
        self.train_X = train_X
        self.train_y = train_y
        
    def train(self, M):
        self.M = M
        X = self.train_X
        y = self.train_y
        N = len(self.train_y)
        w = np.ones(N)/N #initialize weight w
        self.weak_classifiers = []
        self.alpha = []
        for m in range(self.M):
            #fit a classifier to training set using weight w
            clf = DecisionTreeClassifier(max_depth=3)
            clf = clf.fit(X,y,sample_weight=w)
            #compute the error
            accu = clf.score(X,y,sample_weight=w)
            err = 1 - accu
            #compute alpha
            alp = np.log((1-err)/err)
            #reset w
            y_pred = clf.predict(X)
            for i in range(N):
                if y[i] != y_pred[i]:
                    w[i] *= np.exp(alp)
            w /= w.sum()
            self.alpha.append(alp)
            self.weak_classifiers.append(clf)
        return w
            
    def predict(self, test_X):
        N = len(test_X)
        pred_y = np.zeros(N)
        for m in range(self.M):
            pred_y += (self.alpha[m])*(self.weak_classifiers[m].predict(test_X))
        for i in range(N):
            if pred_y[i] > 0:
                pred_y[i] = 1
            else:
                pred_y[i] = -1
        self.pred_y = pred_y
        return self.pred_y
    
    def score(self, test_y):
        pred_y = self.pred_y
        N = len(test_y)
        #calculate the accuracy
        k = 0
        for i in range(N):
            if pred_y[i] == test_y[i]:
                k += 1
        accuracy = float(k)/N
        error = 1 - accuracy
        return error
    
```

### 5.1.3 Plot train and test error as a function of the number of rounds

```python
adb = AdaBoost(train_X, train_y)
train_errors = []
test_errors = []
for rounds in range(1,11):
    adb.train(rounds)
    adb.predict(train_X)
    train_error = adb.score(train_y)
    train_errors.append(train_error)
    adb.predict(test_X)
    test_error = adb.score(test_y)
    test_errors.append(test_error)
    
plt.plot(range(1,11), train_errors)
plt.plot(range(1,11), test_errors)
plt.legend(['Train Errors','Test Errors'])
plt.xlabel('Number of Rounds')
plt.ylabel('Error')
plt.savefig('5_1_3.png', dpi=300)
plt.title("Train and Test Errors with Different Number of Rounds of AdaBoosting")
```

### 5.1.2 Visualize the AdaBoost training procedure

```python
train_y = np.array([-1 if y == 0 else y for y in train_y])
```

```python
# Parameters
n_classes = 2
plot_colors = "by"
plot_step = 0.02

fig = plt.figure(figsize=(12,4))
adb = AdaBoost(train_X, train_y)

for rounds in range(1,11):
    # Train
    w = adb.train(rounds)
    plt.subplot(2, 5, rounds)
    
    # Plot the decision boundary
    x_min = train_X[:, 0].min() - 1
    x_max = train_X[:, 0].max() + 1
    y_min = train_X[:, 1].min() - 1
    y_max = train_X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    
    Z = adb.predict(np.c_[xx.ravel(),yy.ravel()])
    
    Z = Z.reshape(xx.shape)
    for i in range(len(Z)):
        for j in range(len(Z[i])):
            if Z[i][j] == -1:
                Z[i][j] = 0
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.gist_yarg, alpha=0.5)
    cs2 = plt.contour(cs, levels=cs.levels[::2], color='r')
    plt.axis("tight")
    
    
    order = random.sample(range(0,len(train_X)),350)
    tiny_train_X = np.array([train_X[i] for i in order])
    tiny_train_y = np.array([train_y[i] for i in order])
    
    # Plot the training points
    for i, color in zip(range(-1,2,n_classes), plot_colors):
        idx = np.where(train_y == i)
        plt.scatter(train_X[idx, 0], train_X[idx, 1],s=15000*w, marker='+', c=color,
                    cmap=plt.cm.Paired)

    plt.axis("tight")
    
plt.savefig('5_1_2_2.png', dpi=300)
```
